# Merge all AUC and Threshold summary tables

# Set up

```r
main_dir <- "../8_ROC"                               # CHANGE THIS
```

# Pattern match and read for summary AUCs for the relationships

```r
#AUCs

files1 <- list.files(
  path = main_dir,
  pattern = "^AUC_summary.*\\.txt$",       #scan all AUC summary files for each relationship
  recursive = TRUE,
  full.names = TRUE
)

if (length(files1) == 0) stop("No matching AUC_summary*.txt files found")

#Thresholds

files2 <- list.files(
  path = main_dir,
  pattern = "_LR_thresholds_youden_closest\\.csv$",       #scan all thresholds summary files for each relationship
  recursive = TRUE,
  full.names = TRUE
)

if (length(files2) == 0) stop("No matching Threshold files found")
```

# Create the merged file

```r
#AUC
merged1 <- do.call(
  rbind,
  lapply(files1, function(f) {
    df <- read.table(f, sep = "\t", header = TRUE, stringsAsFactors = FALSE, check.names = FALSE)
    df$source_file <- strsplit(basename(f), "_")[[1]][3]
    df <- df[, c("source_file", setdiff(names(df), "source_file"))]       #put relationship labels in first column
    df
  })
)

print(merged1)

#Thresholds
merged2 <- do.call(
  rbind,
  lapply(files2, function(f) {
    df <- read.table(f, sep = "\t", header = TRUE, stringsAsFactors = FALSE, check.names = FALSE)
    # df$source_file <- strsplit(basename(f), "_")[[1]][3]
    # df <- df[, c("source_file", setdiff(names(df), "source_file"))]       #put relationship labels in first column
    df
  })
)

print(merged2)
```

# Save file

```r
write.table(merged1, file = file.path(main_dir, "AUC_CI_DeLong_summary_merged.txt"),
            sep = "\t", quote = FALSE, row.names = FALSE)

write.table(merged2, file = file.path(main_dir, "Threshold_summary_merged.csv"),
            sep = "\t", quote = FALSE, row.names = FALSE)

cat("Merged", length(files1), "files into", file.path(main_dir, "AUC_CI_DeLong_summary_merged.txt"), "\n")
cat("Merged", length(files2), "files into", file.path(main_dir, "Threshold_summary_merged.csv"), "\n")
```
